## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tit = pd.read_csv('data/train.csv')
```

```python
tit.head()
```

```python
tit.info()
```

```python
tit['Sex'].value_counts()
```

## Basic types of plots in Seaborn

```python
sns.factorplot('Survived', data=tit, kind='count')
```

```python
sns.factorplot('Pclass', 'Survived', data=tit)
```

```python
sns.factorplot('Sex', 'Survived', data=tit)
```

## Combining information

```python
sns.factorplot('Sex', 'Survived', hue='Pclass', data=tit)
```

```python
tit['Age'].hist(bins=50)
```

```python
sns.distplot(tit['Age'].dropna())
```

## More complex combinations to get data overlays

```python
fig = sns.FacetGrid(tit, hue='Sex', aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)
oldest = tit['Age'].max()
fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(tit, hue='Pclass', aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)
oldest = tit['Age'].max()
fig.set(xlim=(0, oldest))
fig.add_legend()
```
