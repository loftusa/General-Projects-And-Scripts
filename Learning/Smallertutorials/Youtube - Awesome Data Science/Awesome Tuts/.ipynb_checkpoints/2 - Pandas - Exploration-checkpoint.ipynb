## Initial EDA

**1. Did LE increase from 1960-2013?**

```python
lexp.mean().plot()
```

**2. Which countries had the highest and lowest LE in 1960?**

```python
print(lexp['1960'].idxmax())
print(lexp['1960'].idxmin())
```

**3. What were the highest, lowest, and mean values in 1960?**

```python
print(np.round(lexp['1960'].max()))
print(np.round(lexp['1960'].min()))
print(np.round(lexp['1960'].mean()))
```

**4. What were the highest, lowest, and mean values for year x?**

```python
def high_low_mean(year):
    print('in year', year)
    print(lexp[year].idxmin(), 'had the lowest LE at', np.round(lexp[year].min()))
    print(lexp[year].idxmax(), 'had the highest LE at', np.round(lexp[year].max()))
    print('The mean LE was', np.round(lexp[year].mean()))
```

```python
for i in range(5):
    high_low_mean(str(2000 + i))
    print('\n')
```

**5. Which countries had the lowest and highest gains in LE?**

```python
lexp['gain'] = np.round((lexp['2013'] - lexp['1960']) / lexp['1960'], 2)
```

## Plotting various distributions

```python
lexp.head()
```

```python
lexp.hist(column='1960', bins=20, edgecolor='white')
```

```python
# Compare distributions for years 1960 and 2013 in subplots
f, (ax1, ax2) = plt.subplots(2)
lexp.hist(column='1960', ax=ax1)
lexp.hist(column='2013', color='green', ax=ax2)
```

```python
# Compare two years in the same frame
f, (ax1) = plt.subplots(1, figsize=LARGE_FIGSIZE)
lexp.hist(column='1960', bins=30, ax=ax1)
lexp.hist(column='2013', bins=30, color='blue', ax=ax1)
plt.xlabel('Life Expectancy')
plt.ylabel('Number of Countries')
plt.title('1960 vs 2013')
plt.legend()
```

```python
def compare_hist(year1, year2):
    f, (ax1) = plt.subplots(1, figsize=LARGE_FIGSIZE)
    lexp.hist(column=year1, bins=30, ax=ax1, alpha=0.8)
    lexp.hist(column=year2, bins=30, color='blue', ax=ax1, alpha=0.8)
    plt.xlabel('Life Expectancy')
    plt.ylabel('Number of Countries')
    plt.title('{} vs {}'.format(str(year1), str(year2)))
```

```python
compare_hist('1960', '2013')
```

### Create widget to compare decades!

```python
# use list comprehension to extract decades
years = [str(year) for year in range(1960, 2014) if year%10==0]
```

```python
years
```

```python
interact(compare_hist, year1=years, year2=years)
```

## Kernel Density Estimate (KDE) and Rugplot

### Seaborn -- Optimized for statistical visualization

```python
# Use Seaborn to plot distributions
# Parameters for histogram, kde, and rug can be set true or false

sns.distplot(lexp['2010'].dropna(), hist=False, kde=True, rug=True, bins=25)
```

```python
def sns_compare(year1, year2):
    f, (ax1) = plt.subplots(1, figsize=LARGE_FIGSIZE)
    ax1.set(ylim=(0, 0.050))
    for yr in range(int(year1), int(year2)):
        sns.distplot(lexp[str(yr)].dropna(), hist=False, kde=True, rug=False, bins=25)
    
```

```python
sns_compare('2000', '2010')
```

```python
interact(sns_compare, year1=years, year2=years)
```

### Compare sets of countries

```python
def compare(countries):
    for country in countries:
        t[country].plot(figsize=(20,10))
        print(country, lexp_gain[country])
    plt.legend()
```
