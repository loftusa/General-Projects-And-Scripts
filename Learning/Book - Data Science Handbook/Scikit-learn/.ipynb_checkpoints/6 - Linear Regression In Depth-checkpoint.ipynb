# Linear Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
np.set_printoptions(suppress=True)
```

## Simple Linear Regression

```python
rng = np.random.RandomState(42)

# make a set of x- and y- values. Y values are scattered normally.
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
# Use the LinearRegression estimator to model a line on this data
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x[:, None], y)

plt.plot(x, model.predict(x[:, None]))
plt.scatter(x, y)
```

```python
# Print the predicted slope and intercept of the data
print(model.intercept_)
print(model.coef_)
```

```python
# Show that you can do this in higher dimensions

rng = np.random.RandomState(42)

X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])

model.fit(X, y)

print(model.intercept_)
print(model.coef_)
```

## Basic function regression

**Polynomial basis functions**

```python
# Add 3 degrees of polynomial features to np.array([1, 2, 3])
from sklearn.preprocessing import PolynomialFeatures

# Instantiate a PolynomialFeatures object and make np.array([1, 2, 3])
x = np.array([1, 2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)

# Add the polynomial features to x
poly.fit_transform(x[:, None])
```

### Polynomial features can do a pseudo-linear fit by adding polynomials to each x-value.

```python
# Make a pipeline that adds 7 polynomial features to the data, then performs a linear regression
from sklearn.pipeline import make_pipeline

poly_model = make_pipeline(PolynomialFeatures(7),
                           LinearRegression())


# Make a scatterplot of a sin wave, with values slightly randomized
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + .1 * rng.randn(50)

# Do a linear regression on this data, then plot it
poly_model.fit(x[:, None], y)

xfit = np.linspace(0, 10, 50)
yfit = poly_model.predict(xfit[:, None])

plt.scatter(x, y)
plt.plot(xfit, yfit, c='r');
```

- This does an n-dimensional linear fit with each x-value having polynomial features added to it.

### Gaussian basis functions

- All about fitting a sum of Gaussian bases

```python
from sklearn.base import BaseEstimator, TransformerMixin


# I don't understand how this class works at all, besides the very vague notion that it uses a gaussian distribution instead of polynomials to transform X's
class GaussianFeatures(BaseEstimator, TransformerMixin):
    """Uniformly spaced Gaussian features for one-dimensional input"""
    
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor
        
    @staticmethod
    def _gauss_basis(x, y, width, axis=None):
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))
    
    def fit(self, X, y=None):
        # Create N centers spread along the data range
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self
        
    def transform(self, X):
        return self._gauss_basis(X[:, :, None], self.centers_,
                                 self.width_, axis=1)
```

```python
# But hey it works
gauss_model = make_pipeline(GaussianFeatures(20),
                            LinearRegression())

gauss_model.fit(x[:, None], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10)
```

## Regularization

```python
# Too many gaussian features leads to overfitting

# Couldn't figure out how to change the color palette :(
plt.scatter(x, y)
for i in range(10, 31):
    model = make_pipeline(GaussianFeatures(i),
                          LinearRegression())
    model.fit(x[:, np.newaxis], y)
    sns.lineplot(xfit, model.predict(xfit[:, np.newaxis]),)
```

```python
def basis_plot(model, title=None):
    fig, ax = plt.subplots(2, sharex=True)
    model.fit(x[:, None], y)
    ax[0].scatter(x, y)
    ax[0].plot(xfit, model.predict(xfit[:, None]))
    ax[0].set(xlabel='x', ylabel='y', ylim=(-1.5, 1.5))
    
    if title:
        ax[0].set_title(title)
        
    ax[1].plot(model.steps[0][1].centers_,
               model.steps[1][1].coef_)
    ax[1].set(xlabel='basis location',
              ylabel='coefficient',
              xlim=(0, 10))
    
model = make_pipeline(GaussianFeatures(30),
                      LinearRegression())

basis_plot(model)
```

- Regularization: penalizes large values of model parameters (??)
- So... makes it harder to overfit (somehow)?

### Ridge Regression ($L_2$ regularization)

- Penalizes sum of squares (2-norms) of the model coefficients

```python
from sklearn.linear_model import Ridge
model = make_pipeline(GaussianFeatures(30), 
                      Ridge(alpha=.05))

basis_plot(model, title='Ridge Regression')
```

### Lasso regularization ($L_1$)

- Penalizes sum of absolute values of regression coefficients
- Favors sparse models where possible

```python
from sklearn.linear_model import Lasso
model = make_pipeline(GaussianFeatures(30),
                      Lasso(alpha=0.001))
basis_plot(model, title='Lasso Regression')
```
