# Basics of MPL: line plots

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
from IPython.display import Image
%matplotlib inline
```

```python
# Plot both sin(x) and x**2 on the same graph
x = np.array(range(10))
fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.plot(x, np.sin(x))
```

```python
# Make a fig and axes manually, then plot on it, using MATLAB interface
fig = plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, x**2)

plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.tan(x))
```

```python
# Do the same thing as above using object-oriented interface
fix, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, x**2)
ax[0, 1].plot(x, np.sin(x))
ax[1, 0].plot(x, np.cos(x))
ax[1, 1].plot(x, np.tan(x))
ax[1, 1].plot(x, np.sin(x))
```

```python
# Make a plot with plt with multiple functions on the same figure
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# Make a plot, change xlim and ylim
x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x), '--', label='cos(x)')
plt.plot(x, np.sin(x), 'r', label='sin(x)')
plt.xlim(-1.0, 11.0)
plt.ylim(-1.5, 1.5)

# Change the min/max values of both x and y with one line
plt.axis([0, 10, -1.0, 1.5], 'equal')

plt.title('cos and sin graph')
plt.xlabel('distance around circle (radians)')
plt.ylabel('output value of cos or sin function')
plt.legend()
```

```python
# Put the above graph on to a 4x4 grid
fig, ax = plt.subplots(2, 2, figsize=(14,7), sharex=True, sharey=True)
ax[0,0].plot(x, np.cos(x), '--', label='cos(x)')
ax[0,0].plot(x, np.sin(x), 'r', label='sin(x)')
ax[0,0].set(xlabel='x label', ylabel='y label', title='title')
ax[0,0].legend()
```
