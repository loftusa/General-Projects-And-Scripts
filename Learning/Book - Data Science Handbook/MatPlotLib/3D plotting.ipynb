## 3D Contour Plots

```python
x = np.arange(3)
y = np.arange(3)

np.meshgrid(x, y)
```

```python
# Make a cool 3D contour shape

def f(x, y):
    return np.cos(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = x.copy()

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Make a figure
# Make a 3d axes
# Make a contour3D graph on the axes
# Set labels for x, y, and z

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
# make the above figure, but with a wireframe

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')  # Make a black wireframe
ax.set_title('wireframe plot')  # Set the title
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('surface')
```

## Surface Triangulations

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))

z = f(x, y)
```

```python
# Make a 3d axes
# Make a scatter plot of x, y, z, c=z, with viridis and chnage the linewidth

ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c=z, cmap='viridis', linewidth=0.5)
```

```python
# Make a 3d axes
# Use trisurf
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='Pastel1')
plt.show()
```
