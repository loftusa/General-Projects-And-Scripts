# Using Stylesheets in matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
%matplotlib inline
```

```python
plt.style.use('ggplot')

x = np.array(range(10))
with plt.style.context('seaborn'):
    plt.plot(x, np.sin(x))
```

```python
def hist_and_lines():
    '''
    Make a figure with 2 subplots and an 11x4 figsize
    Make a histogram on the first axis
    Plot 3 points on the second axis
    Make a legend on the second axis in the lower left with 'a', 'b', 'c'
    '''
    fig, ax = plt.subplots(1, 2, figsize=(6, 2))
    ax[0].hist(np.random.randn(1000))
    col = list('rbg')
    for i in range(3):
        ax[1].scatter(np.random.rand(10), np.random.rand(10), c=col.pop())  # Gotta make this not have overlapping colors
    ax[1].legend(['a', 'b', 'c'], loc='lower left', borderpad=.5, labelspacing=0.5)
        

hist_and_lines()
```

```python
with plt.style.context('seaborn'):
    hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

```python
with plt.style.context('bmh'):
    hist_and_lines()
```
