# The Basics

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
from IPython.display import Image
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.savefig('fig.png')
```

```python
Image('fig.png')
```

## Matlab-style

```python
plt.figure()

plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(3, 1, 2)
plt.plot(x, np.cos(x))
```

## Object-oriented style

```python
fig, ax = plt.subplots(2)
ax[0].plot(np.sin(x))
ax[1].plot(np.cos(x))
```
