# Learning to use MPL

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Basic plotting

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

fig.savefig('sin-cos.svg')
```

## Exploring matlab-style interface

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), c='r')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), c='b')
```
