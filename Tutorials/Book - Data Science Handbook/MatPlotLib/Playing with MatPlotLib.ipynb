# Matplotlib from the ground up

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

## Line plots

### Via matlab interface

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1001)
ax.plot(x, np.sin(x))
```

### Via object-oriented interface

```python
fig, ax = plt.subplots(1)
ax.plot(x, np.sin(x))
```

## Multiple plots

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle = 'dashed')
plt.plot(x, x+2, ':r')
```
