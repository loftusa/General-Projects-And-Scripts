```python
import matplotlib.pyplot as plt
import math
import numpy as np
import matplotlib.mlab as mlab

cmu = 5.54
cstd = 2.01
lmu = 4.70
lstd = 1.93
x = np.linspace(cmu - 3*cstd, cmu + 3*cstd, 100)
```

```python
plt.plot(x, mlab.normpdf(x, cmu, cstd))
plt.plot(x, mlab.normpdf(x, lmu, lstd))
plt.legend(['control', 'lucky'])
plt.show()
```

```python
plt.plot(x, mlab.normpdf(x, cmu, (cstd/math.sqrt(35))))
plt.plot(x, mlab.normpdf(x, lmu, (lstd/(math.sqrt(33)))))
plt.legend(['control', 'lucky'])
plt.show()
```
