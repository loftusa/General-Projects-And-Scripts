# Analyzing Grade Trends in my College Career

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Get data

```python
df = pd.read_csv('Transcript.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.dropna()
df.head()
```

```python
# Make conversion dict
grades = {'A': 4.0, 'A-': 3.7, 'B+': 3.3, 'B': 3.0, 'B-': 2.7, 'C+': 2.3, 'C': 2.0, 'C-': 1.7, 'D+': 1.3, 'D': 1.0, 'D-': 0.7, 'F': 0}
```

```python
grade_nums = []
for i in df['GRADE']:
    if i not in grades:
        grade_nums.append('NaN')
    else:
        grade_nums.append(grades[i])

df['GRADE_NUMS'] = np.array(grade_nums).astype('float64')
df
```
