# Testing HDF files

## TRANSFERRING FROM HDF5 TO EXCEL

```python
import pandas as pd
```

### Below: Change 'file location' below to wherever the features_mean file you want to pull data is.

```python
file_location = r'Z:\Associative Conditioning Behavior Videos_2017 to 2018\Alex Loftus - Put your videos here\Ready for TierPsy\WT - paired\Results\2.14.18 N2 Paired ARL 1 min Pre_feat_manual.hdf5'
f = pd.HDFStore(file_location)
means = f.get('features_means')
```

### Below: Change 'name' to the name of the excel file you want to make

If you want to change what folder the excel file goes to, change the file path in 'writer'.

```python
name = 'output'
writer = pd.ExcelWriter(r'C:\Users\roselab\Desktop\Excel-Converter\{}.xlsx'.format(name))
means[['worm_index', 'n_frames', 'bend_count']].to_excel(writer, 'Sheet1')
writer.save()
```
